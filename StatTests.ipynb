```python
import numpy as np
from scipy import stats
import pandas as pd
```

```python
df = pd.read_csv("/content/top_features.csv")
```

```python
count = 0
F = []
M = []

for i in df["count_sessions"]:  # составляем распределения по полу
  if df["sex"][count] == 0:
    F.append(i)
  else:
    M.append(i)
  count+=1

F = np.array(F)
M = np.array(M)
stats.kstest(M,F) #КС тест
```

```python
count_age = pd.DataFrame({
    "count" : df["count_sessions"],
    "age" : df["age_class"]
})
count_age.corr()
```

```python
count = 0
F = []
M = []

for i in df["mean_per_session"]:  # составляем распределения по полу
  if df["sex"][count] == 0:
    F.append(i)
  else:
    M.append(i)
  count+=1

F = np.array(F)
M = np.array(M)
stats.kstest(M,F) #КС тест
```

```python
mean_session_age = pd.DataFrame({
    "mean_session" : df["mean_per_session"],
    "age" : df["age_class"]
})
mean_session_age.corr()
```

```python
count = 0
F = []
M = []

for i in df["mean_per_day"]:  # составляем распределения по полу
  if df["sex"][count] == 0:
    F.append(i)
  else:
    M.append(i)
  count+=1

F = np.array(F)
M = np.array(M)
stats.kstest(M,F) #КС тест
```

```python
mean_sex = pd.DataFrame({
    "mean" : df["mean_per_day"],
    "age" : df["age_class"]
})
mean_sex.corr()
```
